### Defining X and Y variables

```python
X = df.iloc[:, :-1].values
X
```

```python
Y = df.iloc[:, -1].values
Y
```

### Splitting the dataset into train and test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=1234)
```

### Feature Scaling

- We scale our data to make all the features contribute equally to the result

- Fit_Transform - fit is calculating the mean and variance of each of the features present in our data

- We want our test data to be completely new and a surprise set for our model

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

X_train = sc.fit_transform(X_train)

X_test = sc.transform(X_test)
```

### Training 

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(random_state=0)

model.fit(X_train,Y_train)
```

### Predicting whether the new Customer with age & Salary inputs will buy or not

```python
age = int(input("Enter the new customer age: "))
salary = int(input("Enter the new customer salary: "))
newCust = [[age,salary]]

result = model.predict(sc.transform(newCust))

print(result)

if result == 1:
    print("The new customer will buy")
else:
    print("The new customer won't buy")
```

## Prediction for all test data

```python
Y_pred = model.predict(X_test)
print(np.concatenate((Y_pred.reshape(len(Y_pred),1), Y_test.reshape(len(Y_test),1)),1))
```

### Evaluating model - Confusion matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score

cm = confusion_matrix(Y_test, Y_pred)
print("Confusion matrix: ")
print(cm)

print("The accracy of the model is: {0}%".format(accuracy_score(Y_test,Y_pred)*100))
```
